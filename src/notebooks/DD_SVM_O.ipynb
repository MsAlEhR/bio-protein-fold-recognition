## Classification of Protein Fold using SVM and Ocurrence Features

Read raw data from CSV file.

```python
import csv

data_file = open('./dataset/DD_raw.csv')
csv_data = list(csv.reader(data_file, delimiter=','))
csv_data
```

Extract features using Ocurrence method

```python
import bio_pre

data_FV = bio_pre.occurence_FE(csv_data[1:])
data_FV
```

Convert dataset to Pandas dataframe

```python
import pandas as pd

# Feature names
amino_acides = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N',
                'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']

data_frame = pd.DataFrame(data_FV, columns=['Fold Name', 'Protein Name'] + amino_acides)
```

```python
data_frame.head()
```

Convert class lables to unique integers

```python
data_frame.insert(1, 'Class Labels', pd.factorize(data_frame['Fold Name'])[0])
data_frame.head()
```

Create a training set from pandas data frame

```python
train_data = data_frame[amino_acides].values
labels = data_frame['Class Labels'].values

print("Number of samples: %d, Number of features: %d" % (train_data.shape[0], train_data.shape[1]))
```

### SVM classifier

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

# Hyperparameters
kernel = 'rbf'

# An instance of SVM classifier
svm_cl = SVC(kernel=kernel)
```

Define range of parameters for Grid Search

```python
# Penalty parameter
c_range = {'C': [float(2**i) for i in range(-5, 6)]}
# Gamma parameter for RBF kernel
gamma_range = {'gamma': [float(2**i) for i in range(-10, 3)]} if kernel == 'rbf' else {}

param_range = {**c_range, **gamma_range}

# Arguments for grid search
cv_fold = 10
n_workers = 2 # Number of CPU threads

result = GridSearchCV(svm_cl, param_range, cv=cv_fold, n_jobs=n_workers, refit=False,
                      verbose=1)
```

```python
result.fit(train_data, labels)
```

Best results

```python
print("Best accuracy: %.2f" % (result.best_score_ * 100))
```
